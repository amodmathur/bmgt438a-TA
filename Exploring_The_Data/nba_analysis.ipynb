```python
# Dataset obtained from https://github.com/sivabalanb/Data-Analysis-with-Pandas-and-Python/blob/master/nba.csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('nba_data.csv')
```

```python
# Descriptive Statistics
df.describe()
```

```python
df['Salary'].describe()
```

```python
df['College'].describe()
```

```python
# Histogram
nba_ages = df['Age'].plot.hist()
nba_ages.set(xlabel='Ages', title='NBA Players Age Distribution')
```

```python
# Bar Graphs
# notice the use of the value_counts() function to transform categorical data into numeric representation of its # of observations
nba_positions = df['Position'].value_counts().plot.barh()
nba_positions.set(xlabel='Frequency', ylabel='Position', title='NBA Positions Distribution')
```

```python
# Kernel Density Estimates
df['Age'].plot.kde()
```

```python
df['Salary'].plot.kde()
```

```python
# Boxplots
boxplot_salary = df.boxplot(column=['Salary'], by=['Position'])
```

```python
# Scatter Plots
plt.scatter(df["Age"], df["Salary"],  alpha=0.5)
plt.title("Salary Versus Age")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
# Correlation Matrix
df[["Salary","Age"]].corr()
```
